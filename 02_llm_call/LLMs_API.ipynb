### Einfache Anfrage an Gemini-Modell

1. OpenAI-API-Key erstellen auf der [API-Plattform](https://ai.google.dev/gemini-api/docs/api-key)
2. Beispielcode findet sich auf der [API-Dokumentation unter "Overview"](https://ai.google.dev/gemini-api/docs)
3. Eine Übersicht über verfügbare Modelle und die korrespondierenden Kosten findet sich [hier](https://ai.google.dev/gemini-api/docs/models)
4. Die offizielle Kurzanleitung inden Sie [hier](https://ai.google.dev/gemini-api/docs/quickstart?hl=de)

```python
from google import genai

client = genai.Client()

response = client.models.generate_content(
    model="gemini-3-flash-preview",
    contents="schreib mir eine Zusammenfassung über die Vor- und Nachteile von OOP in Python",
)

print(response.text)
```

### Einfache Anfrage an OpenAI-Modell 

1. OpenAI-API-Key erstellen auf der [API-Plattform](https://platform.openai.com/home)
2. Beispielcode findet sich auf der [OpenAI Developers Plattform](https://developers.openai.com/api/docs/quickstart?language=python)
3. Eine Übersicht über verfügbare Modelle und die korrespondierenden Kosten findet sich [hier](https://developers.openai.com/api/docs/pricing)

```python
from openai import OpenAI
client = OpenAI()

response = client.responses.create(
    model="gpt-5.4",
    input="schreib mir eine Zusammenfassung über die Vor- und Nachteile von OOP in Python"
)

print(response.output_text)
```
