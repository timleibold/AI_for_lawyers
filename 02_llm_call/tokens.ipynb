### Tokenisierung von Text

Die offizielle Dokumentation zur library ```tiktoken``` finden Sie [hier](https://developers.openai.com/cookbook/examples/how_to_count_tokens_with_tiktoken)

wie viele Token hat ein Text?

````python
import tiktoken

encoding = tiktoken.get_encoding("cl100k_base") # nach Encoding name
num_tokens = len(encoding.encode("Bundesverfassungsgericht"))
print(num_tokens)
````

````python
encoding = tiktoken.encoding_for_model("gpt-4o") # nach Model name
indexes = encoding.encode("Bundesverfassungsgericht")
print(indexes)
print(len(indexes))
````

````python
token_encoded = encoding.decode(indexes)
print(token_encoded)
````

````python
token_list = [encoding.decode_single_token_bytes(token) for token in indexes]
token_list # das b vor den tokens zeigt an, dass es sich um byte strings handelt
````

````python
for token in token_list:
    print(type(token)) # Entfernt das b vor den byte strings
````
